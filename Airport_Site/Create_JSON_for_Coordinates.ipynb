```python
import pandas as pd


data = pd.read_csv("large_airports_lat_long.csv")
```

```python
data.head()
```

```python
data = data.rename(columns={"latitude_deg":"lat", "longitude_deg":"long"})
```

```python
data = data.drop(data.columns[0], axis=1)
data
```

```python
json_data = []

for item in data:
    airport = {
        "iata_code":data["iata_code"],
        "name": data["name"],
        "latitude":data["lat"],
        "longitude":data["long"]
    }
    json_data.append(airport)
```

```python
data_dict = data.to_dict(orient='index')
```

```python
data_dict
```

```python
import json
# Specify the path to save the JSON file
json_file_path = "large_airport_data.json"

# Write the dictionary to the JSON file
with open(json_file_path, 'w') as json_file:
    json.dump(data_dict, json_file, indent=4)
```
